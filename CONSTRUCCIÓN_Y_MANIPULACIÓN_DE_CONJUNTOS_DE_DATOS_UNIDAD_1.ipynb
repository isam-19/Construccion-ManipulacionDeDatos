```
import pandas as pd

# Cargar y leer el archivo CSV desde Colab
df_csv = pd.read_csv('Tarifas_aplicadas_de_Gas_Natural_20250401.csv')

# Cargar el archivo JSON desde la URL de Datos Abiertos
url_json = "https://www.datos.gov.co/resource/duwb-pa7h.json"
df_json = pd.read_json(url_json)
```

```
# Verificación de la correcta carga de los datasets
# Ver las primeras filas de ambos datasets para revisar sus estructuras

print("Primeras filas del CSV:")
print(df_csv.head())

print("\nPrimeras filas del JSON:")
print(df_json.head())
```

# **Análisis por separado**:

---

Inicialmente se hará el análisis por separado y posterior a ello se combinarán los Datasets en uno unico.

```
# Información general del CSV
print("\nInformación del DataFrame CSV:")
print(df_csv.info())

# Información general del JSON
print("\nInformación del DataFrame JSON:")
print(df_json.info())

# Descripción estadística del CSV
print("\nDescripción estadística del CSV:")
print(df_csv.describe())

# Descripción estadística del JSON
print("\nDescripción estadística del JSON:")
print(df_json.describe())

# Comprobar valores faltantes en el CSV
print("\nValores faltantes en el CSV:")
print(df_csv.isnull().sum())

# Comprobar valores faltantes en el JSON
print("\nValores faltantes en el JSON:")
print(df_json.isnull().sum())

# Comprobar si hay valores negativos en el CSV, solo en columnas numéricas
print("\nValores negativos en el CSV:")
for column in df_csv.select_dtypes(include=['number']):  # Iterar solo por columnas numéricas
    print(f"Columna '{column}': {(df_csv[column] < 0).sum()}")  # Imprimir resultados por columna

# Comprobar si hay valores negativos en el JSON, solo en columnas numéricas
print("\nValores negativos en el JSON:")
for column in df_json.select_dtypes(include=['number']):  # Iterar solo por columnas numéricas
    print(f"Columna '{column}': {(df_json[column] < 0).sum()}")  # Imprimir resultados por columna

# Verificar duplicados en el CSV
print("\nDuplicados en el CSV:")
print(df_csv.duplicated().sum())

# Verificar duplicados en el JSON
print("\nDuplicados en el JSON:")
print(df_json.duplicated().sum())
```

# **Análisis de los datos combinados**:  
---
Se hace el ejercicio nuevamente con los datos de ambos datasets en uno único combinado


```
# Combinar los datos de ambos DataFrames en un único DataFrame
df = pd.concat([df_csv, df_json], axis=0, ignore_index=True)
```

```
# Visualizar las primeras filas del DataFrame combinado
print("Primeras filas del DataFrame combinado:")
print(df.head())
```

```
# Información general sobre el DataFrame
print("\nInformación del DataFrame combinado:")
print(df.info())

# Descripción estadística del DataFrame
print("\nDescripción estadística del DataFrame combinado:")
print(df.describe())

# Comprobación de valores faltantes
print("\nValores faltantes en el DataFrame combinado:")
print(df.isnull().sum())

# Comprobación de duplicados
print("\nNúmero de filas duplicadas en el DataFrame combinado:")
print(df.duplicated().sum())

# Comprobación de valores negativos en columnas numéricas
# Para detectar valores negativos, primero identificaremos identificamos las columnas numéricas
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

print("\nValores negativos en las columnas numéricas del DataFrame combinado:")
for col in numerical_columns:
    negative_values = (df[col] < 0).sum()
    print(f"{col}: {negative_values} valores negativos")
```
